# Titles, labels, lagends

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]
x2 = [9, 5, 2, 4, 4]
y2 = [6, 8,  7, 4, 1]

plt.plot(x,y)
plt.plot(x2, y2)
plt.title("Tech With Rushbah")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]
x2 = [9, 5, 2, 4, 4]
y2 = [6, 8,  7, 4, 1]

plt.plot(x,y)
plt.plot(x2, y2)
plt.title("Tech With Rushbah")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]
x2 = [9, 5, 2, 4, 4]
y2 = [6, 8,  7, 4, 1]

plt.plot(x,y, label="Line 1")
plt.plot(x2, y2, label="Line 2")
plt.title("Tech With Rushbah")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]
x2 = [9, 5, 2, 4, 4]
y2 = [6, 8,  7, 4, 1]

plt.plot(x,y, label="Line 1")
plt.plot(x2, y2, label="Line 2")
plt.title("Tech With Rushbah")
plt.legend()
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]
x2 = [9, 5, 2, 4, 4]
y2 = [6, 8,  7, 4, 1]

plt.plot(x,y, label="Line 1", color="salmon")
plt.plot(x2, y2, label="Line 2", color="black")
plt.title("Tech With Rushbah")
plt.legend()
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
plt.show()
```

## Bar Charts

```python
import matplotlib.pyplot as plt

x = [1, 5 ,6, 9, 10]
y = [5, 6, 4, 5, 4]

plt.bar(x,y, label="Line 1", color="red")
plt.title("Tech With Rushbah")
plt.legend()
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt

day = [21, 32, 69, 56, 85, 45, 50, 30, 10, 15, 16, 42, 66, 77, 11, 2, 32, 9, 91, 85, 81, 92, 27]
place = [x for x in range(len(day))]

plt.hist(day,place, histtype="bar", label="This is label")
plt.title("Tech With Rushbah")
plt.legend()
plt.xlabel("this is the x label")
plt.ylabel("this is the y label")
plt.show()
```
